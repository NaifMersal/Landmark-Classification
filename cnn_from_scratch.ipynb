<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__

 **CUDA OUT MEMORY** This is the most frequent message I have seen. Since I was training on GTX 970 (4GB VRAM about 800MB is taken by system -> total 3.2 GB). These are the steps I have taken to tackle that.

First I decided to use a VGG like CNN but then I can't fit it into memory, so I build a smaller one and got about 20% accuracy. I added attention layers the accuracy raised to 24% with a size of 500MB cause of Linear Layers(more parameters less grad so it could fit).  
 I added connections and reduce filters which will be called bottlenecks later. I get an accuracy of 46% and a size of 300MB. However, there was my limit there. I was wondering how models like mobile net have a size of 15MB, so I had to take a look at the [mobile net](https://arxiv.org/abs/1704.04861) paper, they introduced the separable convolutional layers. I made, in general, 5 versions(actual 25 versions) using separable convolutional layers(SConv). All these layers start with the regular Conv layer since we will half the size of the input, we need to get the best combination for each filter.

- V1 was able to get 51% with size of 222MB it was VVG-like but uses SConv layers and has so many Linear layers.
- V2 was able to get about 48% with a size of 105MB similar to V1 but has fewer Linear layers and more SConv layers
- V3 was able to get about 51% with a size of 91MB similar to V2 but has fewer Linear layers and more SConv layers


- V4 was able to get about 59% with a size of 14MB I added the Bottlenecks and connections 
- V5 was able to get about 62% with a size of 2.1MB made it deeper, used AVGpooling, used only one Linear Layer, and used regular Conv in the last two layers so we can get the best combinations for each filter. In fact, the last Conv layer is somehow Linear.

````python
# NOTE: you might need to restart the notebook before running this step
# If you get an error about RuntimeError: Can't redefine method: forward on class
# restart your notebook then execute only this cell
from src.predictor import Predictor
from src.helpers import compute_mean_and_std
from src.model import MyModel
from src.data import get_data_loaders
import torch

data_loaders = get_data_loaders(batch_size=1)

# First let's get the class names from our data loaders
class_names = data_loaders["train"].dataset.classes

# Then let's move the model_transfer to the CPU
# (we don't need GPU for inference)
model = MyModel(num_classes=50, dropout=0.5).cpu()

# Let's make sure we use the right weights by loading the
# best weights we have found during training
# NOTE: remember to use map_location='cpu' so the weights
# are loaded on the CPU (and not the GPU)

model.load_state_dict(torch.load('checkpoints/k_V5_val_loss.pt'))

# Let's wrap our model using the predictor class
mean, std = compute_mean_and_std()
predictor = Predictor(model, class_names, mean, std).cpu()

# Export using torch.jit.script
scripted_predictor =torch.jit.script(predictor) 

scripted_predictor.save("checkpoints/original_exported.pt")
````
